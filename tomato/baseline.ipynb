```python
xgb_res = []
rf_res = []


avg_1_seed = [2]
# [0.0454]
# [0.0420]
avg_75_seed = [67]
# [0.0461]
# [0.0435]
avg_50_seed = [3]
# [0.0480]
# [0.0436]
avg_25_seed = [54]
# [0.0489]
# [0.0442]
avg_167_seed = [0]
# [0.0485]
# [0.046]
avg_125_seed = [79]
# [0.0488]
# [0.0453]


for seed in range(100):
    data_subset = data.sample(frac=1/8, random_state=seed)

    train_X = data_subset[FEATURE_COLS].values
    train_y = data_subset[LABEL_COLS].values

    val_indices, train_indices = util.get_fold_indices_rand(len(train_X), 1, FOLDS)

    xgb_maes = []
    rf_maes = []

    for fold, (fold_train_indices, fold_val_indices) in enumerate(zip(train_indices, val_indices)):
        xgb_regressor = xgb.XGBRegressor(tree_method='hist', seed=42)
        rf_regressor = SklearnRF(random_state=42)

        fold_train_X, fold_val_X = train_X[fold_train_indices], train_X[fold_val_indices]
        fold_train_y, fold_val_y = train_y[fold_train_indices], train_y[fold_val_indices]

        # Fitting the model
        xgb_regressor.fit(fold_train_X, fold_train_y.ravel())
        rf_regressor.fit(fold_train_X, fold_train_y.ravel())

        # Predict the model
        xgb_pred = xgb_regressor.predict(fold_val_X)
        rf_pred = rf_regressor.predict(fold_val_X)

        xgb_mae = MAE(fold_val_y, xgb_pred)
        rf_mae = MAE(fold_val_y, rf_pred)
        xgb_maes.append(xgb_mae)
        rf_maes.append(rf_mae)

        # print("MAE : % f" % mae)

    # print("Mean MAE : % f" % np.mean(mae))
    xgb_res.append(xgb_maes)
    rf_res.append(rf_maes)

xgb_res = np.array(xgb_res)
rf_res = np.array(rf_res)

print(np.median(xgb_res))
print(np.median(rf_res))
```

```python
xgb_mean = xgb_res.mean(axis=1)
rf_mean = rf_res.mean(axis=1)
```
